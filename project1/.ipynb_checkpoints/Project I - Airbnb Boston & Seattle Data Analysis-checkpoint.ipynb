##### Data Understanding

`Boston Airbnb dataset`
1. There are **3,585** unique listings in Boston in the year **from 2016-09-06 to 2017-09-05**.
1. Avaregely, the listings have **221 days available** for rent throughout the 365 days.
1. There are **3,585 rows and 95 columns** for the Boston listings dataset.
1. 'square_feet' has more than 95% missing values; 'has_availability', 'license', and 'jurisdiction_names' are empty columns. 
1. 'price' data type should be **numerical**.
1. There are **68,275** rows of reviews for **2,829** unique Boston listings.
1. To answer our business questions, we don't need detailed content in the Boston reviews dataset. 

`Seattle Airbnb dataset`
1. There are **3,818** unique listings in Seattle in the year **from 2016-01-04 to 2017-01-02**.
1. Avaregely, the listings have **251 days available** for rent throughout the 365 days.
1. There are **3,818 rows and 92 columns** for the Seattle listing dataset; 'access', 'interaction', and 'house_rules' are in the Boston listings dataset only.
1. 'square_feet' has more than 95% missing values; 'license' is an empty column.
1. 'price' data type should be **numerical**.
1. There are **84,849** rows of reviews for **3,191** unique Seattle listings.
1. To answer our business questions, we don't need the detailed content in the Seattle reviews dataset.

##### Question 3: How well can we predict the price with a Linear Regression model?

I will select features that are most relevant to listing prices based on the second question, and fit a linear regression model for Airbnb Seattle and Airbnb Boston respectively.

```python
#feature selection
boston_var = ['security_deposit', 'cleaning_fee', 'accommodates', 'bathrooms', 'beds',
             'property_type', 'room_type', 'bed_type', 'cancellation_policy', 'host_response_time', 
              'price']
seattle_var = ['security_deposit', 'cleaning_fee', 'accommodates', 'bathrooms', 'beds',
             'property_type', 'room_type', 'bed_type', 'cancellation_policy', 'host_response_time', 
              'price']
```

```python
#deal with missing values
boston_df = df_listing_clean[df_listing_clean['market']=='boston'][boston_var]
seattle_df = df_listing_clean[df_listing_clean['market']=='seattle'][seattle_var]
display(boston_df.info())
display(seattle_df.info())
```

`Deal with Missing Values`:
1. security_deposit in both datasets have many missing values, I will **drop the column** directly.
2. missing values in cleaning_fee can be **replaced with mean value**.
3. othere columns have little missing values, I will **drop the rows** with missing values directly.

```python
boston_df.drop('security_deposit', axis=1, inplace=True)
seattle_df.drop('security_deposit', axis=1, inplace=True)
```

```python
display(seattle_df.describe())
display(boston_df.describe())
```

```python
seattle_df['cleaning_fee'] = seattle_df['cleaning_fee'].fillna(seattle_df['cleaning_fee'].mean())
boston_df['cleaning_fee'] = boston_df['cleaning_fee'].fillna(boston_df['cleaning_fee'].mean())
```

```python
seattle_df.dropna(how='any', inplace=True)
boston_df.dropna(how='any', inplace=True)
```

Below is a function used to create `dummy variables` for dataframe with categorical values. I learnt it from the [Udacity Data Sicence Course](https://www.udacity.com/course/data-scientist-nanodegree--nd025).

```python
#create dummy variables
def create_dummy_df(df, cat_cols, dummy_na):
    '''
    INPUT:
    df - pandas dataframe with categorical variables you want to dummy
    cat_cols - list of strings that are associated with names of the categorical columns
    dummy_na - Bool holding whether you want to dummy NA vals of categorical columns or not
    '''
    for col in cat_cols:
        try:
            dummy_df = pd.get_dummies(df[col], dummy_na=dummy_na, prefix=col, prefix_sep='_', drop_first=True)
            df.drop(col, axis=1, inplace=True)
            df = pd.concat([df, dummy_df], axis=1)
        except:
            continue       
    return df
```

```python
seattle_df_mol = create_dummy_df(seattle_df, 
                             ['property_type','room_type', 'bed_type', 'cancellation_policy','host_response_time'],
                             dummy_na=False)
boston_df_mol = create_dummy_df(boston_df, 
                             ['property_type','room_type', 'bed_type', 'cancellation_policy','host_response_time'],
                             dummy_na=False)
```
